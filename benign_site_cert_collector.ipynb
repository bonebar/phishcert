```python
import socket
import ssl
import OpenSSL
import pandas as pd
import numpy as np
import json
import re
import cryptography
from cryptography import x509
from cryptography.hazmat.backends import default_backend
from cryptography.x509.extensions import Extensions
from cryptography.x509 import GeneralName
from cryptography.x509 import oid
from cryptography.x509.oid import ExtensionOID
from cryptography.x509.oid import NameOID
from cryptography.x509 import PolicyInformation
import urllib.request
from urllib.parse import * 
import time
import statistics
```

```python
#Top Websites from Majectic Million - https://majestic.com/reports/majestic-million
df = pd.read_csv (r'ProcessedBenignSites.csv')
top = df['Domain'].tolist()
topsites = top[0:100]
print(topsites)
```

```python
#Import processed benign websites txt file
benign = open("bengignProcessed.txt", "r")
topSitesProcessed = benign.read()
topSitesProcessed = topSitesProcessed.split('\n')
benign.close()
topsites = topSitesProcessed[10000:]
```

```python
startCell = time.time()
certlist = []
timeperiod = []

for site in topsites:
    #startCode = time.time()
    print(str(topsites.index(site))+ ': ' + site)
    sitelist = []
    sitelist.append(site)
    
    try:
        #Creates connection and collects certificate
        conn = ssl.create_connection((site, 443))
        #sock.settimeout(5)
        context = ssl.SSLContext(ssl.PROTOCOL_SSLv23)
        sock = context.wrap_socket(conn, server_hostname=site)
        pem = ssl.DER_cert_to_PEM_cert(sock.getpeercert(True))
        pem = bytes(pem, 'utf-8')
        sitelist.append(pem)
        cert = x509.load_pem_x509_certificate(pem, default_backend())
    except:
        continue
        
    # Retrieves Subject
    subjectCN = cert.subject.get_attributes_for_oid(NameOID.COMMON_NAME)
    subjectO = cert.subject.get_attributes_for_oid(NameOID.ORGANIZATION_NAME)
    if subjectO != []:
        #print(subjectO[0].value)
        sitelist.append(subjectO[0].value) 
    else:
        sitelist.append('')

    if subjectCN != []:
        #print(subjectCN[0].value)
        sitelist.append(subjectCN[0].value)
    else:
        sitelist.append('')

    # Retrieves Issuer
    issuerCN = cert.issuer.get_attributes_for_oid(NameOID.COMMON_NAME)
    issuerO = cert.issuer.get_attributes_for_oid(NameOID.ORGANIZATION_NAME)
    if issuerO != []:
        #print(issuerO[0].value)
        sitelist.append(issuerO[0].value)
    else:
        sitelist.append('')

    if issuerCN != []:
        #print(issuerCN[0].value)
        sitelist.append(issuerCN[0].value)
    else:
        sitelist.append('')

    # Retrieves Not Valid Before
    try:
        validstart = str(cert.not_valid_before)
        #print(validstart)
        sitelist.append(validstart)
    except:
        sitelist.append('')
        
    # Retrieves Not Valid After
    try:
        validend = str(cert.not_valid_after)
        #print(validend)
        sitelist.append(validend)
    except:
        sitelist.append('')
        
    # Retrieves Subject Alternative Names
    try:
        ext_SAN = cert.extensions.get_extension_for_oid(ExtensionOID.SUBJECT_ALTERNATIVE_NAME)
        full_SAN = ext_SAN.value.get_values_for_type(x509.DNSName)
        #print(full_SAN)
        sitelist.append(full_SAN)
    except:
        sitelist.append([])
        
    # Retrieves policy OID to determine validation level
    try:
        policy = cert.extensions.get_extension_for_oid(ExtensionOID.CERTIFICATE_POLICIES)
        for p in policy.value:
            OID = p.policy_identifier.dotted_string

            if OID == '2.23.140.1.2.1':
                #print('DV')
                sitelist.append('DV')
                break

            elif OID == '2.23.140.1.2.2':
                #print('OV')
                sitelist.append('OV')
                break

            elif OID == '2.23.140.1.2.3':
                #print('IV')
                sitelist.append('IV')
                break

            elif OID == '2.23.140.1.1':
                #print('EV')
                sitelist.append('EV')
                break

            else:
                continue

        if len(sitelist) != 10:
            sitelist.append('')  
    except:
       sitelist.append('')  
    certlist.append(sitelist)
#     endCode = time.time() - startCode
#     timeperiod.append(endCode)
#     print(endCode)
elapsed = time.time() - startCell
print(elapsed)
```

```python
#Output data into .csv file
columns = ['Domain', 'PEM', 'Subject O', 'Subject CN', 'Issuer O', 'Isssuer CN', 'Validity Start', 'Validity End', 'SAN', 'Validation Level']
df = pd.DataFrame(np.array(certlist, dtype="object"), columns=columns)
name = "BenignSitesP2"
file_name = name + '.csv'
df.to_csv(file_name, index=False, header=True)
df
```
